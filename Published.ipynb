```python
import matplotlib.pyplot as plt
import numpy as np

# plt.rcParams.update({'font.size': 14})
# Sample data for scatter plots
x_values = [[2, 4, 6, 12], [2, 4, 6, 12], [2, 4, 6, 12], [2, 4, 6, 12], [2, 4, 6, 12]]
y_values = [
    [72.1, 41.6, 23.3, -8.5],
    [75.4, 46.7, 26.6, -4.5],
    [30.2, 50.3, 29.9, -2.9],
    [14.6, 26, 35.8, -1.8],
    [5.0, 4.5, 5.8, 11.4],
]

y_values_2 = [
    [29.3,14.5,9.6,4.0],
    [17.6,9.4,6.1,2.5],
    [10.5,4.9,3.3,1.3],
    [9.2,6.5,2.9,1.0],
    [3.6,2.3,3.1,1.0],
]

# Positions of the vertical dashed lines
vlines_positions = [1.8, 2.0, 4.0, 6.0, 12.0]

# Create subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 10))

# Titles for each subplot
titles = ["(a) 360 days", "(b) 180 days", "(c) 90 days", "(d) 60 days", "(e) 30 days"]

# Define common y-axis limits
ylim_min = -1
ylim_max = 1  # Adjust according to your data

# Plot data for each subplot
for idx in range(5):
    row = idx // 2
    col = idx % 2
    ax = axs[row, col]
    ax.scatter(x_values[idx], (100 + np.array(y_values[idx])) / 100 - 1, color="black")
    ax.scatter(x_values[idx], (100 + np.array(y_values_2[idx])) / 100 - 1, color="black")
    ax.set_xlabel("Survey Frequency, times/yr", fontsize=14)
    ax.set_ylabel("Emission deviation", fontsize=14)
    ax.tick_params(axis="y", labelsize=10)
    ax.tick_params(axis="x", labelsize=10)

    lns1 = ax.plot(x_values[idx], (100 + np.array(y_values[idx])) / 100 - 1, linestyle="-", color="orange", label="Repair")
    lns2 = ax.plot(x_values[idx], (100 + np.array(y_values_2[idx])) / 100 - 1, linestyle="-", color="blue", label="Non-repair")

    # Add red dashed line at y = 0
    ax.axhline(y=0, color='gray', linestyle='--')

    # Add vertical dashed line at specified x position
    ax.axvline(x=vlines_positions[idx], color='darkgreen', linestyle='--')

    # Set y-axis limits
    ax.set_ylim(ylim_min, ylim_max)
    ax.set_title(titles[idx], fontsize=14)

    # Create legend for each subplot
    ax.legend(loc="upper right", fontsize=12)

# Hide the empty subplot
fig.delaxes(axs[2, 1])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Provided data
x_values = [[360, 180, 90, 60, 30], [360, 180, 90, 60, 30], [360, 180, 90, 60, 30], [360, 180, 90, 60, 30]]

# Calculate 25th and 75th percentiles
percentile_25 = [
    [19.6,7.9,-6.0,-10.5,-13.5], [10.4,4.8,0.9,-4.4,-10.9], [6.6,3.5,-2.2,-0.0,-7.3], [2.6,1.2,-0.0,-0.4,-0.3]]
percentile_75 = [
    [39.4,27.4,28.5,28.9,26.0], [19.4,14.2,9.8,16.0,17.8], [12.3,9.3,3.1,5.7,13.7], [5.4,3.9,2.7,2.5,2.4]]

# Convert to numpy arrays for easier manipulation
percentile_25 = np.array(percentile_25)
percentile_75 = np.array(percentile_75)

# Calculate the interquartile range (IQR)
difference = (100 + (percentile_75 - percentile_25)) / 100 -1

# Plotting
plt.figure(figsize=(10, 6))
labels = ["Semiannually", "Quarterly", "Bimonthly", "Monthly"]
for i in range(difference.shape[0]):
    plt.plot(x_values[i], difference[i], marker='o', label=labels[i])

plt.xlabel('Emission duration, days', fontsize=14)
plt.ylabel('Interquartile range', fontsize=14)
plt.ylim(0, 0.6)  # Set the y-axis limits
plt.legend(fontsize=12)
plt.show()
```
