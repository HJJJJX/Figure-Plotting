```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Specific x and y values
x_values = ['30', '60', '90', '180', '360']
y_values = ['12', '6', '4', '2']

# Example data corresponding to each (x, y) pair
data0 = np.array([
    [11.4, -1.8, -2.9, -4.5, -8.5],
    [5.8, 35.8, 29.9, 26.6, 23.3],
    [4.5, 26, 50.3, 46.7, 41.6],
    [5.0, 14.6, 30.2, 75.4, 72.1]
])

data = (100 + np.array(data0))/100

# Define custom colormap
colors = [(0, 'darkred'), (0.5, 'white'), (1, 'darkblue')]
cmap = LinearSegmentedColormap.from_list('custom', colors)

# Set custom width and height
width = 10
height = 6

# Plot heatmap with adjusted aspect ratio
plt.figure(figsize=(width, height))  # Set figure size
plt.imshow(data, cmap=cmap, interpolation='nearest', aspect='auto', vmin=0, vmax=2)
plt.colorbar()
plt.title('Uncertainty Heatmap for Repair Scenarios')

# Adjust x and y ticks to be in the middle of each grid box
plt.xticks(np.arange(len(x_values)), x_values)
plt.yticks(np.arange(len(y_values)), y_values)

# Add data values inside each block
for i in range(len(y_values)):
    for j in range(len(x_values)):
        plt.text(j, i, '{:.1f}'.format(data[i, j]), ha='center', va='center', color='black')

# Add x and y axis titles
plt.xlabel('Emission duration, days')
plt.ylabel('Survey frequency, times/yr')

plt.tight_layout()
plt.savefig('Heatmap1.png')
plt.show()
```
