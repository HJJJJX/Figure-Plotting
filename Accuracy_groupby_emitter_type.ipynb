```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x_labels = ['X1', 'X2', 'X3', 'X4', 'X5']
group1 = [4, 7, 1, 8, 5]
group2 = [2, 5, 2, 6, 3]
group3 = [3, 6, 3, 7, 4]

x = np.arange(len(x_labels))  # the label locations
width = 0.2  # the width of the bars

# Create the plot
fig, ax = plt.subplots()

# Plot each group
rects1 = ax.bar(x - width, group1, width, label='Group 1')
rects2 = ax.bar(x, group2, width, label='Group 2')
rects3 = ax.bar(x + width, group3, width, label='Group 3')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_xlabel('X Values')
ax.set_ylabel('Scores')
ax.set_title('Scores by group and X values')
ax.set_xticks(x)
ax.set_xticklabels(x_labels)
ax.legend()

# Function to add a value label on each bar
def add_labels(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

# Add labels to the bars
add_labels(rects1)
add_labels(rects2)
add_labels(rects3)

# Show the plot
plt.tight_layout()
plt.show()
```
