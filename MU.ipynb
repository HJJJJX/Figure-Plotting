```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for scatter plots
x_values = [[2, 4, 6, 12], [2, 4, 6, 12], [2, 4, 6, 12], [2, 4, 6, 12], [2, 4, 6, 12]]
y_values = [
    [29.3,14.5,9.6,4.0],
    [17.6,9.4,6.1,2.5],
    [10.5,4.9,3.3,1.3],
    [9.2,6.5,2.9,1.0],
    [3.6,2.3,3.1,1.0],
]

y_values_2 = [
    [29.1,14.5,9.4,4.3],
    [17.5,10.2,6.8,2.6],
    [9.1,5.2,3.2,1.3],
    [8.2,5.7,2.9,1.0],
    [3.0,1.6,2.6,0.9],
]


# Create subplots
fig, axs = plt.subplots(3, 2, figsize=(10, 10))

# Titles for each subplot
titles = ["(a) 360 days", "(b) 180 days", "(c) 90 days", "(d) 60 days", "(e) 30 days"]

# Define common y-axis limits
ylim_min = -0.25
ylim_max = 0.3  # Adjust according to your data

# Plot data for each subplot
for j in range(3):
    for k in range(2):
        idx = j * 2 + k
        if idx == 5:  # Skip the 6th subplot
            continue
        ax = axs[j, k]
        ax.scatter(x_values[idx], (100 + np.array(y_values[idx]))/100-1, color="black")
        ax.scatter(x_values[idx], (100 + np.array(y_values_2[idx]))/100-1, color="black")
        ax.set_xlabel("Survey Frequency, times/yr")
        ax.set_ylabel("Emission Ratio")
        ax.tick_params(axis="y")

        ax.plot(x_values[idx], (100 + np.array(y_values[idx]))/100-1, linestyle="-", color="blue", label="w/o measurement uncertainty")
        ax.plot(x_values[idx], (100 + np.array(y_values_2[idx]))/100-1, linestyle="-", color="orange", label="w measurement uncertainty")

        # Add red dashed line at y = 1.0
        ax.axhline(y=0, color='red', linestyle='--')

        # Set y-axis limits
        ax.set_ylim(ylim_min, ylim_max)
        ax.set_title(titles[idx])

        # Create legend for each subplot
        ax.legend(loc="lower right")

# Hide the empty subplot
fig.delaxes(axs[2, 1])

plt.tight_layout()
plt.show()
```
